##### Q.2) Write a program in python to perform following task :
##### Standardizing Data (transform them into a standard Gaussian distribution with a mean of 0 and a standard deviation of 1) (Use winequality-red.csv)

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

df = pd.read_csv("./wineQuality.csv")

df.head()

s = df['quality']

mu , sd = np.mean(s) , np.std(s)
s = np.random.normal(mu,sd,len(s))

print("Mean :" , np.mean(s))
print("Standard Deviation :" , np.std(s))

plt.figure(figsize=(10,7))
plt.hist(s , 30)
plt.ylabel("quality")
plt.xlabel("alcohole")
plt.title("Histogram of Wine Quality")
plt.show()
```
