##### Q.2 B) Write a Python program to find the maximum and minimum value of a given flattened array

```python
import numpy as np
arr = [
    [0,1],
    [2,3]
]

flatt_arr  = np.array(arr).flatten()
print("Minimum :" ,flatt_arr.min())
print("Maximum :" , flatt_arr.max())
```
